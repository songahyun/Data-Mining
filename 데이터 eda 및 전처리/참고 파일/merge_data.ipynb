```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 파일 경로 리스트
file_paths = [
    "2021_delete_boarding_f.csv",
    "2022_delete_boarding_f.csv",
    "2023_delete_boarding_f.csv"
]

# 빈 데이터프레임 리스트 생성
dataframes = []

# 각 파일 불러오기
for file in file_paths:
    df = pd.read_csv(file)

    # 2023년도 데이터에서 '유형' 열 제거
    if '유형' in df.columns:
        df = df.drop(columns=['유형'])

    # 리스트에 추가
    dataframes.append(df)

# 모든 데이터프레임을 합치기
combined_df = pd.concat(dataframes, ignore_index=True)

# 합쳐진 데이터 확인
print(combined_df.head())

# 합쳐진 데이터를 CSV 파일로 저장
combined_df.to_csv("merged_boarding.csv", index=False)
```

```python
# 데이터 불러오기
file_path = "merged_boarding.csv"
data = pd.read_csv(file_path)

data
```

```python
# '승차거리'의 평균값 계산
mean_distance = data['승차거리'].mean()

# 평균값 미만인 경우만 필터링
filtered_data = data[data['승차거리'] < mean_distance]

# 결과 확인
print(filtered_data)

# CSV 파일로 저장
filtered_data.to_csv('merged_filtered_data.csv', index=False)
```
